```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://www.stat4decision.com/wine_data.csv", index_col = 0)
df["type"].value_counts()
```

```python
df.info()
```

```python
y=df["type"]
x=df.drop("type",axis=1)
x_train, x_test, y_train, y_test  = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

modele_rf = RandomForestClassifier(
    n_estimators=100,                # Nombre d'arbres dans la forêt (100 par défaut)
    criterion='gini',                # Critère pour mesurer la qualité des divisions ('gini' par défaut)
    max_depth=None,                  # Profondeur maximale des arbres (illimitée par défaut)
    min_samples_split=2,             # Nombre minimum d'échantillons nécessaires pour diviser un nœud
    min_samples_leaf=1,              # Nombre minimum d'échantillons dans une feuille
    min_weight_fraction_leaf=0.0,    # Fraction minimale de poids dans une feuille
    max_features='sqrt',             # Nombre de caractéristiques à considérer pour diviser un nœud ('sqrt' recommandé)
    max_leaf_nodes=None,             # Nombre maximal de feuilles (illimité par défaut)
    min_impurity_decrease=0.0,       # Gain minimal requis pour effectuer une division
    bootstrap=True,                  # Utilisation d'échantillonnage bootstrap pour construire les arbres
    oob_score=False,                 # Score hors sac désactivé
    n_jobs=None,                     # Nombre de threads (utilise 1 par défaut, ou -1 pour utiliser tous les cœurs)
    random_state=None,               # Graine pour rendre les résultats reproductibles (facultatif)
    verbose=0,                       # Niveau de verbosité (aucun affichage par défaut)
    warm_start=False,                # Si True, réutilise les arbres existants pour ajouter de nouveaux
    class_weight=None,               # Pondération des classes (None par défaut)
    ccp_alpha=0.0,                   # Coût-complexité pour la taille de l'arbre (aucune taille par défaut)
    max_samples=None                 # Nombre d'échantillons pour le bootstrap (tous les échantillons par défaut)
    
)
```

```python
modele_rf.fit(x_train, y_train)
```

```python
pd.DataFrame(modele_rf.feature_importances_,
              index = x_train.columns, 
              columns = ["importance"]).sort_values(
     "importance", 
     ascending = False)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(f"Le pourcentage de bien classés est de : {accuracy_score(y_test, modele_rf.predict(x_test))*100} %")
```

```python
pd.DataFrame(confusion_matrix(y_test, modele_rf.predict(x_test)),
             index = ["blanc_données", "rouge_données"],
             columns = ["blanc_predit", "rouge_predit"])
```
